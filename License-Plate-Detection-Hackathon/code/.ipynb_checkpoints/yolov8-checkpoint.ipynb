```python
import boto3

# Initialize S3 client
s3 = boto3.client('s3')

# Define S3 bucket and object key
bucket_name = 'license-plate-detection-hackathon'
object_key = 'License Plate Detector.v2i.yolov8.zip'  # Ensure this matches exactly

# Define local file path
local_file_path = 'dataset.zip'

try:
    # Download the dataset from S3
    s3.download_file(bucket_name, object_key, local_file_path)
    print(f"File downloaded successfully to {local_file_path}")
except Exception as e:
    print(f"Error: {e}")
```

```python
import zipfile
import os

# Define the path to the zip file and extraction directory
zip_file_path = 'dataset.zip'
extraction_path = './Dataset/'

# Create extraction directory if it doesn't exist
if not os.path.exists(extraction_path):
    os.makedirs(extraction_path)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"Extracted files to {extraction_path}")
```

```python
import boto3
s3 = boto3.client('s3')

# Define S3 bucket and object key
bucket_name = 'license-plate-detection-hackathon'
object_key = 'ezyzip.zip'  # Ensure this matches exactly

# Define local file path
local_file_path = 'ultralytics.zip'

try:
    # Download the dataset from S3
    s3.download_file(bucket_name, object_key, local_file_path)
    print(f"File downloaded successfully to {local_file_path}")
except Exception as e:
    print(f"Error: {e}")
```

```python
import zipfile
import os

zip_file_path = 'ultralytics.zip'
extraction_path = './ultralytics/'

# Create extraction directory if it doesn't exist
if not os.path.exists(extraction_path):
    os.makedirs(extraction_path)

try:
    # Extract the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extraction_path)
    print(f"Extracted files to {extraction_path}")
except FileNotFoundError:
    print(f"File {zip_file_path} not found.")
except zipfile.BadZipFile:
    print(f"File {zip_file_path} is not a zip file or is corrupted.")
except Exception as e:
    print(f"An error occurred: {e}")
```
